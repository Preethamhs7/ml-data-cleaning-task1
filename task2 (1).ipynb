```python
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()

```

```python
df.hist(bins=30, figsize=(15, 10), color='skyblue')
plt.suptitle("Distribution of Numeric Features", fontsize=16)
plt.show()
```

```python
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()

sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
```

```python
df.skew(numeric_only=True)
```

```python
Observation

* Age and Fare show presence of outliers.
* Fare is positively correlated with survival; Pclass negatively.
* Survived class shows imbalance.
* Visuals show women and upper-class passengers had higher survival rates.
```
